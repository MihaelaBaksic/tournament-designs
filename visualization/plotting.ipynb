## Visualization and plotting of measurement results for Latin Suqares and Tournament Designs implementation

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Rotation Latin square construction algorithm

```python
def relative_time(t0, tn):
    return tn / t0
```

```python
ls_rotation = pd.read_csv('../measurements/latin_square.csv', dtype={'parameter': int, 'construction_time': float})

ls_rotation_none = ls_rotation[ls_rotation['optimization'] == 'none'].copy()
ls_rotation_O = ls_rotation[ls_rotation['optimization'] == 'O'].copy()
ls_rotation_O1 = ls_rotation[ls_rotation['optimization'] == 'O1'].copy()
ls_rotation_O2 = ls_rotation[ls_rotation['optimization'] == 'O2'].copy()
ls_rotation_O3 = ls_rotation[ls_rotation['optimization'] == 'O3'].copy()

t_none = ls_rotation_none['construction_time'].iloc[0]
ls_rotation_none['relative_time'] = ls_rotation_none.apply(lambda row: relative_time(t_none, row['construction_time']), axis=1)

t_O = ls_rotation_O['construction_time'].iloc[0]
ls_rotation_O['relative_time'] = ls_rotation_O.apply(lambda row: relative_time(t_O, row['construction_time']), axis=1)

t_O1 = ls_rotation_O1['construction_time'].iloc[0]
ls_rotation_O1['relative_time'] = ls_rotation_O1.apply(lambda row: relative_time(t_O1, row['construction_time']), axis=1)

t_O2 = ls_rotation_O2['construction_time'].iloc[0]
ls_rotation_O2['relative_time'] = ls_rotation_O2.apply(lambda row: relative_time(t_O2, row['construction_time']), axis=1)

t_O3 = ls_rotation_O3['construction_time'].iloc[0]
ls_rotation_O3['relative_time'] = ls_rotation_O3.apply(lambda row: relative_time(t_O3, row['construction_time']), axis=1)


```

```python
fig, ax = plt.subplots(1)
plt.plot( ls_rotation_none['parameter'], ls_rotation_none['construction_time'], label='none')
plt.plot( ls_rotation_O['parameter'], ls_rotation_O['construction_time'], label='O')
plt.plot( ls_rotation_O1['parameter'], ls_rotation_O1['construction_time'], label='O1')
plt.plot( ls_rotation_O2['parameter'], ls_rotation_O2['construction_time'], label='O2')
plt.plot( ls_rotation_O3['parameter'], ls_rotation_O3['construction_time'], label='O3')

ax.set_ylabel('execution_time [s]')
ax.set_xlabel('n')
plt.title('Latin square basic algorithm')
plt.legend(title='Compiler optimization')

plt.savefig('latin_square_basic_algorithm.jpg', dpi=500)
```

### Latin square complete algorithm

```python
ls_complete = pd.read_csv('../measurements/latin_square_complete.csv', dtype={'parameter': int, 'construction_time': float})

ls_complete_none = ls_complete[ls_complete['optimization'] == 'none'].copy()
ls_complete_O = ls_complete[ls_complete['optimization'] == 'O'].copy()
ls_complete_O1 = ls_complete[ls_complete['optimization'] == 'O1'].copy()
ls_complete_O2 = ls_complete[ls_complete['optimization'] == 'O2'].copy()
ls_complete_O3 = ls_complete[ls_complete['optimization'] == 'O3'].copy()


t_none = ls_complete_none['construction_time'].iloc[0]
ls_complete_none['relative_time'] = ls_complete_none.apply(lambda row: relative_time(t_none, row['construction_time']), axis=1)

t_O = ls_complete_O['construction_time'].iloc[0]
ls_complete_O['relative_time'] = ls_complete_O.apply(lambda row: relative_time(t_O, row['construction_time']), axis=1)

t_O1 = ls_rotation_O1['construction_time'].iloc[0]
ls_complete_O1['relative_time'] = ls_complete_O1.apply(lambda row: relative_time(t_O1, row['construction_time']), axis=1)

t_O2 = ls_complete_O2['construction_time'].iloc[0]
ls_complete_O2['relative_time'] = ls_complete_O2.apply(lambda row: relative_time(t_O2, row['construction_time']), axis=1)

t_O3 = ls_complete_O3['construction_time'].iloc[0]
ls_complete_O3['relative_time'] = ls_complete_O3.apply(lambda row: relative_time(t_O3, row['construction_time']), axis=1)

```

```python
fig, ax = plt.subplots(1)
plt.plot( ls_complete_none['parameter'], ls_complete_none['construction_time'], label='none')
plt.plot( ls_complete_O['parameter'], ls_complete_O['construction_time'], label='O')
plt.plot( ls_complete_O1['parameter'], ls_complete_O1['construction_time'], label='O1')
plt.plot( ls_complete_O2['parameter'], ls_complete_O2['construction_time'], label='O2')
plt.plot( ls_complete_O3['parameter'], ls_complete_O3['construction_time'], label='O3')

ax.set_ylabel('execution_time [s]')
ax.set_xlabel('n')
plt.title('Complete Latin square construction algorithm')
plt.legend(title='Compiler optimization')

plt.savefig('latin_square_complete_algorithm.jpg', dpi=500)
```

### Latin square k-construction algorithm

```python
ls_kconstruction = pd.read_csv('../measurements/latin_square_kconstruction.csv', dtype={'parameter': int, 'construction_time': float})

ls_kconstruction_none = ls_kconstruction[ls_kconstruction['optimization'] == 'none'].copy()
ls_kconstruction_O = ls_kconstruction[ls_kconstruction['optimization'] == 'O'].copy()
ls_kconstruction_O1 = ls_kconstruction[ls_kconstruction['optimization'] == 'O1'].copy()
ls_kconstruction_O2 = ls_kconstruction[ls_kconstruction['optimization'] == 'O2'].copy()
ls_kconstruction_O3 = ls_kconstruction[ls_kconstruction['optimization'] == 'O3'].copy()

t_none = ls_kconstruction_none['construction_time'].iloc[0]
ls_kconstruction_none['relative_time'] = ls_kconstruction_none.apply(lambda row: relative_time(t_none, row['construction_time']), axis=1)

t_O = ls_kconstruction_O['construction_time'].iloc[0]
ls_kconstruction_O['relative_time'] = ls_kconstruction_O.apply(lambda row: relative_time(t_O, row['construction_time']), axis=1)

t_O1 = ls_kconstruction_O1['construction_time'].iloc[0]
ls_kconstruction_O1['relative_time'] = ls_kconstruction_O1.apply(lambda row: relative_time(t_O1, row['construction_time']), axis=1)

t_O2 = ls_kconstruction_O2['construction_time'].iloc[0]
ls_kconstruction_O2['relative_time'] = ls_kconstruction_O2.apply(lambda row: relative_time(t_O2, row['construction_time']), axis=1)

t_O3 = ls_kconstruction_O3['construction_time'].iloc[0]
ls_kconstruction_O3['relative_time'] = ls_kconstruction_O3.apply(lambda row: relative_time(t_O3, row['construction_time']), axis=1)
```

```python
fig, ax = plt.subplots(1)
plt.plot( ls_kconstruction_none['parameter'], ls_kconstruction_none['construction_time'], label='none')
plt.plot( ls_kconstruction_O['parameter'], ls_kconstruction_O['construction_time'], label='O')
plt.plot( ls_kconstruction_O1['parameter'], ls_kconstruction_O1['construction_time'], label='O1')
plt.plot( ls_kconstruction_O2['parameter'], ls_kconstruction_O2['construction_time'], label='O2')
plt.plot( ls_kconstruction_O3['parameter'], ls_kconstruction_O3['construction_time'], label='O3')

ax.set_ylabel('execution_time [s]')
ax.set_xlabel('n')
plt.title('Latin square k construction algorithm')
plt.legend(title='Compiler optimization')

plt.savefig('latin_square_k_construction_algorithm.jpg', dpi=500)
```
